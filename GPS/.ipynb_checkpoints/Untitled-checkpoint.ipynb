### import

```python
import numpy as np
import matplotlib.pyplot as plt
```

## showData

```python
di = None
with open('GPSData.txt') as f:
    di = f.read().split('\n')
    
d = np.zeros((len(di)-1, 2))
for i in range(len(di)-1):
    s = di[i].split(',')
    d[i, 0] = float(s[0])
    d[i, 1] = float(s[1])
    #print('{0},{1}' .format(d[i, 0],d[i, 1]))

plt.xlim(d[:,0].min(), d[:,0].max())
plt.ylim(d[:,1].min(), d[:,1].max())
plt.scatter(d[:,0], d[:,1])
plt.show()
```
