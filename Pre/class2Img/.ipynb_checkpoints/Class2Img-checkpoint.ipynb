```python
import numpy as np
import time
import matplotlib.pyplot as plt
%matplotlib inline
from keras.datasets import mnist
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.optimizers import Adam
from matplotlib import cm
from PIL import Image, ImageDraw

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(60000, 784)
x_train = x_train.astype('float32')
x_test = x_test.reshape(10000, 784)
x_test = x_test.astype('float32')
y_train_raw = y_train
y_test_raw = y_test
y_train = np_utils.to_categorical(y_train, 10)
y_test = np_utils.to_categorical(y_test, 10)
y_train = y_train.astype('float32')
y_test = y_test.astype('float32')
#x_train = (x_train - x_train.min()) / (x_train.max() - x_train.min())
#x_test = (x_test - x_test.min()) / (x_test.max() - x_test.min())
x_train /= 255 # max = 1?
x_test /= 255
x_train_o = np.zeros((10, 784))
y_train_o = np.zeros((10, 10))
x_train_o = x_train[[1, 3, 5, 7, 2, 0, 13, 15, 17, 19]]
y_train_o = y_train[[1, 3, 5, 7, 2, 0, 13, 15, 17, 19]]
print(y_train_o)
```

```python
plt.gray()
for id in range(3):
    plt.subplot(1, 3, id + 1)
    img = x_train[id].reshape(28, 28)
    #img = x_train[id, :, :]
    plt.imshow(255 - img)
    #plt.text(24.5, 26, "%d" % y_train[id],
    #         color='cornflowerblue', fontsize=18)
    #plt.grid('on', color='white')
    plt.axis('off')
plt.show()
```

```python
model = Sequential()
model.add(Dense(256, input_dim=10, activation='relu'))
#model.add(Dense(64, activation='relu'))
#model.add(Dense(128, activation='relu'))
#model.add(Dense(256, activation='relu'))
model.add(Dense(784, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer=Adam(), metrics=['accuracy'])
```

```python
model.fit(y_train_o, x_train_o, epochs=1000, batch_size=1000,
         verbose=0, validation_data=(y_train_o, x_train_o))
```

```python
import matplotlib.pyplot as plt
image_height = 28
image_width = 28
import keras.models

encoder = keras.models.clone_model(model)
encoder.compile(optimizer='adam',
                    loss='binary_crossentropy')
encoder.set_weights(model.get_weights())
# 最終段のレイヤーを取り除く
#encoder.pop()


#p = np.random.random_integers(0, len(x_test), 10)
p = [1, 3, 5, 7, 2, 0, 13, 15, 17, 19]
x_test_sampled = x_train[p]
y_test_sampled = y_train[p]
print(y_train_raw[p])
y_test_sampled_pred = model.predict_proba(y_test_sampled,
                                              verbose=0)
#y_test_sampled_enc = encoder.predict_proba(y_test_sampled,
#                                               verbose=0)

fig, axes = plt.subplots(2, 10)
for i in range(10):
    # 元画像を上段に表示する
    img = x_test_sampled[i].reshape(image_height, image_width)
    axes[0][i].imshow(img, cmap=cm.gray_r)
    axes[0][i].axis('off')
    #axes[0][i].set_title(label, color='red')
    # AutoEncoder で次元圧縮した画像を下段に表示する
    #enc_img = y_test_sampled_enc[i].reshape(6, 6)
    #axes[1][i].imshow(enc_img, cmap=cm.gray_r)
    #axes[1][i].axis('off')
    #AutoEncoder で復元した画像を下段に表示する
    pred_img = y_test_sampled_pred[i].reshape(image_height, image_width)
    axes[1][i].imshow(pred_img, cmap=cm.gray_r)
    axes[1][i].axis('off')

plt.show()
```

```python
n = 50

result = np.zeros((10, n, 784))
for i in range(n):
    result[0, i] = model.predict_proba(np.array([[1-i/n, i/n, 0., 0., 0., 0., 0., 0., 0., 0.]]), verbose=0)
for i in range(n):
    result[1, i] = model.predict_proba(np.array([[0, 1-i/n, i/n, 0., 0., 0., 0., 0., 0., 0.]]), verbose=0)
for i in range(n):
    result[2, i] = model.predict_proba(np.array([[0, 0, 1-i/n, i/n, 0., 0., 0., 0., 0., 0.]]), verbose=0)
for i in range(n):
    result[3, i] = model.predict_proba(np.array([[0, 0, 0., 1-i/n, i/n, 0., 0., 0., 0., 0.]]), verbose=0)
for i in range(n):
    result[4, i] = model.predict_proba(np.array([[0, 0, 0., 0., 1-i/n, i/n, 0., 0., 0., 0.]]), verbose=0)
for i in range(n):
    result[5, i] = model.predict_proba(np.array([[0, 0, 0., 0., 0., 1-i/n, i/n, 0., 0., 0.]]), verbose=0)
for i in range(n):
    result[6, i] = model.predict_proba(np.array([[0, 0, 0., 0., 0., 0., 1-i/n, i/n, 0., 0.]]), verbose=0)
for i in range(n):
    result[7, i] = model.predict_proba(np.array([[0, 0, 0., 0., 0., 0., 0., 1-i/n, i/n, 0.]]), verbose=0)
for i in range(n):
    result[8, i] = model.predict_proba(np.array([[0, 0, 0., 0., 0., 0., 0., 0., 1-i/n, i/n]]), verbose=0)
for i in range(n):
    result[9, i] = model.predict_proba(np.array([[i/n, 0, 0., 0., 0., 0., 0., 0., 0, 1-i/n]]), verbose=0)
    
images = []
#fig, axes = plt.subplots(10, n)
for a in range(10):
    for i in range(n):
        # 元画像を上段に表示する
        img = result[a, i].reshape(image_height, image_width)
        imgc = Image.fromarray(np.uint8(255-img*255)).convert('P').resize((255, 255))
        images.append(imgc)

        #axes[a][i].imshow(images[a*n+i], cmap=cm.gray_r)
        #axes[a][i].axis('off')

plt.show()
images[0].save('test.gif',
               save_all=True, append_images=images[1:], optimize=False, duration=40, loop=0)
```

## cat&dog

```python
imgs = []
imgs.append(np.array(Image.open('cat.jpg').resize((192, 108))))
imgs.append(np.array(Image.open('dog.jpg').resize((192, 108))))
imgs = np.array(imgs).astype('float32')/255
print(imgs.shape)
imgs = imgs.reshape(2, 108*192*3)
print(imgs.shape)

y_train_c = np_utils.to_categorical([0, 1], 2)
print(y_train_c)

plt.imshow(imgs[0].reshape(108, 192, 3))
plt.show()
```

```python
model = Sequential()
model.add(Dense(1024, input_dim=2, activation='relu'))
model.add(Dense(62208, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer=Adam(), metrics=['accuracy'])
```

```python
model.fit(y_train_c, imgs, epochs=100, batch_size=1000,
         verbose=1, validation_data=(y_train_c, imgs))
```

```python
import matplotlib.pyplot as plt
image_height = 108
image_width = 192
import keras.models

encoder = keras.models.clone_model(model)
encoder.compile(optimizer='adam',
                    loss='binary_crossentropy')
encoder.set_weights(model.get_weights())
# 最終段のレイヤーを取り除く
#encoder.pop()


#p = np.random.random_integers(0, len(x_test), 10)
p = [0,1]
x_test_sampled = imgs[p]
y_test_sampled = y_train_c[p]
print(y_train_raw[p])
y_test_sampled_pred = model.predict_proba(y_test_sampled,
                                              verbose=0)
#y_test_sampled_enc = encoder.predict_proba(y_test_sampled,
#                                               verbose=0)
n = 2
fig, axes = plt.subplots(2, n)
for i in range(n):
    # 元画像を上段に表示する
    img = x_test_sampled[i].reshape(image_height, image_width, 3)
    axes[0][i].imshow(img, cmap=cm.gray_r)
    axes[0][i].axis('off')
    #axes[0][i].set_title(label, color='red')
    # AutoEncoder で次元圧縮した画像を下段に表示する
    #enc_img = y_test_sampled_enc[i].reshape(6, 6)
    #axes[1][i].imshow(enc_img, cmap=cm.gray_r)
    #axes[1][i].axis('off')
    #AutoEncoder で復元した画像を下段に表示する
    pred_img = y_test_sampled_pred[i].reshape(image_height, image_width, 3)
    axes[1][i].imshow(pred_img, cmap=cm.gray_r)
    axes[1][i].axis('off')

plt.show()
```

```python
n = 50
image_height=108
image_width=192

result = np.zeros((2, n, 62208))
for i in range(n):
    result[0, i] = model.predict_proba(np.array([[1-i/n, i/n]]), verbose=0)
for i in range(n):
    result[1, i] = model.predict_proba(np.array([[i/n, 1-i/n]]), verbose=0)

    
images = []
fig, axes = plt.subplots(2, n)
for a in range(2):
    for i in range(n):
        # 元画像を上段に表示する
        img = result[a, i].reshape(image_height, image_width, 3)
        imgc = Image.fromarray(np.uint8(255-img*255)).convert('P').resize((255, 255))
        images.append(imgc)

        axes[a][i].imshow(images[a*n+i], cmap=cm.gray_r)
        axes[a][i].axis('off')

plt.show()
images[0].save('test.gif',
               save_all=True, append_images=images[1:], optimize=False, duration=40, loop=0)
```
