### import

```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
readDatas = None
with open('WeatherData.txt') as f:
    readDatas = f.read().split('\n')

timeAxis = np.zeros(len(readDatas)-1)
pressureAxis = np.zeros(len(readDatas)-1)
tempAxis = np.zeros(len(readDatas)-1)
humidityAxis = np.zeros(len(readDatas)-1)
i = 0
for readData in readDatas:
    if readData != '':
        datas = readData.split(',')
        dte = datetime.datetime.strptime(datas[3], '%Y:%m:%d:%H:%M')
        timeAxis[i] = int(datas[4])
        pressureAxis[i] = float(datas[0])
        tempAxis[i] = float(datas[1])
        humidityAxis[i] = float(datas[2])
        i+=1

fig = plt.figure(figsize=(20,20))
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(timeAxis, pressureAxis)
ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(timeAxis, tempAxis)
ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(timeAxis, humidityAxis)
plt.show()
```
