# MIDI Test

```python
import pretty_midi
import numpy as np

result = ''
# MIDIファイルのロード
#midi_data = pretty_midi.PrettyMIDI('test.mid')
midi_data = pretty_midi.PrettyMIDI('Liszt_lacampanella.mid')
# トラック別で取得
midi_tracks = midi_data.instruments
# トラック１のノートを取得
notes = midi_tracks[0].notes
export = np.zeros((len(notes), 4))
i = 0
for note in notes:
    # ベロシティー、ノートナンバー、
    # ノートオンタイム、ノートオフタイム
    # の順でノート情報が渡される
    export[i, 0] = note.start
    export[i, 1] = note.end
    export[i, 2] = note.pitch
    export[i, 3] = note.velocity
    result += str(note.start) + ',' + str(note.end) + ',' + str(note.pitch) + ',' + str(note.velocity) + '\n'
    print(note)
    i+=1
    
print(export)
print(result)
np.savetxt('midi2csv.csv', export, fmt='%.5f')
```
